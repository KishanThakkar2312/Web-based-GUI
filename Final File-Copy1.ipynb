```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import numpy as np
import plotly
import plotly.graph_objects as go
import dash_daq as daq
from collections import deque
import pandas as pd
from dash.dependencies import Output, Input
import dash_bootstrap_components as dbc
from dash import State


df = pd.read_csv("Trial_file.csv")
  
X = deque(maxlen = 50)
X.append(0)
X.append(1)
  
Y = deque(maxlen = 50)
Y.append(df["Temeprature"][0])
Y.append(df["Temeprature"][1])

Y1 = deque(maxlen = 50)
Y1.append(df["Humidity"][0])
Y1.append(df["Humidity"][1])


Y2 = deque(maxlen = 50)
Y2.append(df["CO2"][0])
Y2.append(df["CO2"][1])


Y3 = deque(maxlen = 50)
Y3.append(df["Misc"][0])
Y3.append(df["Misc"][1])


app = dash.Dash(external_stylesheets=[dbc.themes.CYBORG])


app.layout = html.Div([
    
    # This is a header container having title and node selection
    html.Div([ 
        
        # This div element is defined to create a row inside a header container
        html.Div([
                    # Column for heading
                    html.Div([
                        
                        html.H5(["Air Quality Dashboard"])
                        
                            ], className = "ml-5 col-3 ml-2 p-1 text-center"),
                    
                    #column for node selection
                    html.Div([
                        
                        
                        dbc.DropdownMenu(
                                            label="Node",
                                            menu_variant="dark",
                                            color="secondary",
                                            #size="lg",
                                            children=[
                                                dbc.DropdownMenuItem("Node 1"),
                                                dbc.DropdownMenuItem("Node 2"),
                                                dbc.DropdownMenuItem("Node 3"),
                                            ])                        
                            ],className = "  col-2 p-1 text-white font-weight-bold text-center"),
                
            
                ],className= "row p-2 row justify-content-around align-items-center bg-light text-white "),
        
        
            ], className = "container-fluid"),
    
    #End of header container
    
    
    #Body container-1
    html.Div([
        
        # Main row element having two main containers
        html.Div([
            
            # Graphs container
            html.Div([
                
                
                # Row 1 of graph container
                html.Div([
                    
                    # 1st graph column
                    html.Div([
                        
                        html.P(["Temperature"]),
                        
                        html.Div([
                                dcc.Graph(id='temp-graph', animate = True),
                                dcc.Interval(
                                        id = 'graph-update',
                                        interval = 2000,
                                        n_intervals = 0 ),
                           ]),
                        
                            ],className = "col mr-2 mt-2 bg-light graphs"),# end of 1st graph column
                    
                    # 2nd graph column
                    html.Div([
                        
                        html.P(["Humidity"]),
                        dcc.Graph(
                                  
                                id='hum-graph'
                            
                                    ),
                        
                            ],className = "col mt-2 bg-light graphs")#end of 2nd graph column

                    
                    
                        ],className="row"), # End of row-1 of graph container
                
                    # Row 2 of graph container
                    html.Div([
                    
                    # 3rd graph column   
                    html.Div([
                        
                        html.P(["CO2"]),
                        dcc.Graph(
                                  
                                id='co2-graph'
                            
                                    ),
                        
                            ],className = "col mr-2 mt-2 bg-light graphs"),
                    
                    #4th graph column 
                    html.Div([
                        
                        html.P(["Misc Gases"]),
                        dcc.Graph(
                                  
                                id='misc-graph'
                            
                                    ),
                        
                            ],className = "col mt-2 bg-light graphs")

                    
                    
                        ],className="row")# end of Row 2 of graph container

                
                
                
                    ],className = "col-10 text-light"),# End of graph container
            
            
            # Gauge charts container
            html.Div([
                
                #gauge temperature chart
                 html.Div([
                        
                        html.H5(["Temperature"]),
                        
                        dcc.Graph(
                                  
                                id='gtemp-graph'
                            
                                    ),
                        
                            ],className = "row bg-light gauge"),
                html.Div([
                        
                        html.H5(["Humidity"]),
                        
                        dcc.Graph(
                                  
                                id='ghum-graph'
                            
                                    ),
                        
                            ],className = "row bg-light gauge"),
                
                html.Div([
                        
                        html.H5(["CO2"]),
                        
                        dcc.Graph(
                                  
                                id='gco2-graph'
                            
                                    ),
                        
                            ],className = "row bg-light gauge"),
                
                html.Div([
                        
                        html.H5(["Misc Gases"]),
                        
                        dcc.Graph(
                                  
                                id='gmisc-graph'
                            
                                    ),
                        
                            ],className = "row  bg-light gauge"),
                


          
                
                    ],className = "col mt-2 gaugecol")
            
            
            
            
            
            ],className = "row pl-2")# End of main row element of body container-1
        
        
        
        
        
        
        
        
    ],className = "container-fluid ")# End of body container-1
    
    #Container for Daily average
    
    
    
    
])

@app.callback(
    Output('temp-graph', 'figure'),
    [ Input('graph-update', 'n_intervals') ]
)
  
def update_graph_scatter(n):
    X.append(n)
    Y.append(df["Temeprature"][n])
    
  
    data =  go.Scatter(
            x=list(X),
            y=list(Y),
            name='Scatter',
            mode= 'lines+markers'
    )
  
    return {'data': [data],
            'layout' : go.Layout(
                                    xaxis=dict(range=[min(X),max(X)]),
                                    yaxis=dict(range=[31,33]),
                                    margin=dict(l=20, r=20, t=20, b=20),
                                    template ="plotly_dark",
                                    xaxis_title="Time",
                                    yaxis_title="Temp(°C)",
                        
                                )}

@app.callback(
    Output('hum-graph', 'figure'),
    [ Input('graph-update', 'n_intervals') ]
)
  
def update_graph_scatter(n):
    X.append(n)
    Y1.append(df["Humidity"][n])
    
  
    data =  go.Scatter(
            x=list(X),
            y=list(Y1),
            name='Scatter',
            mode= 'lines+markers',
            marker=dict(
                color='#FF0000')
        
    )
  
    return {'data': [data],
            'layout' : go.Layout(
                                    xaxis=dict(range=[min(X),max(X)]),
                                    yaxis=dict(range=[68,72]),
                                    margin=dict(l=20, r=20, t=20, b=20,pad=10),
                                    template ="plotly_dark",
                                    xaxis_title="Time",
                                    yaxis_title="Humidity",
                        
                                )}


@app.callback(
    Output('co2-graph', 'figure'),
    [ Input('graph-update', 'n_intervals') ]
)
  
def update_graph_scatter(n):
    X.append(n)
    Y2.append(df["CO2"][n])
    
  
    data =  go.Scatter(
            x=list(X),
            y=list(Y2),
            name='Scatter',
            mode= 'lines+markers',
            marker=dict(
                color='#FFFF00')
        
    )
  
    return {'data': [data],
            'layout' : go.Layout(
                                    xaxis=dict(range=[min(X),max(X)]),
                                    yaxis=dict(range=[400,900]),
                                    margin=dict(l=20, r=20, t=20, b=20,pad=10),
                                    template ="plotly_dark",
                                    xaxis_title="Time",
                                    yaxis_title="CO2",
                        
                                )}


@app.callback(
    Output('misc-graph', 'figure'),
    [ Input('graph-update', 'n_intervals') ]
)
  
def update_graph_scatter(n):
    X.append(n)
    Y3.append(df["Misc"][n])
    
  
    data =  go.Scatter(
            x=list(X),
            y=list(Y3),
            name='Scatter',
            mode= 'lines+markers',
            marker=dict(
                color='#008000')
        
    )
  
    return {'data': [data],
            'layout' : go.Layout(
                                    xaxis=dict(range=[min(X),max(X)]),
                                    yaxis=dict(range=[400,650]),
                                    margin=dict(l=20, r=20, t=20, b=20,pad=10),
                                    template ="plotly_dark",
                                    xaxis_title="Time",
                                    yaxis_title="Misc Gases",
                        
                                )}


    
@app.callback(
    Output('gtemp-graph', 'figure'),
    [ Input('graph-update', 'n_intervals') ]
)
  
def update_graph_scatter(n):

    Y = df["Temeprature"][n]
    
    fig = go.Figure(go.Indicator(
    mode = "gauge+number",
    value = Y,
    domain = {'x': [0, 1], 'y': [0, 1]},
    title = {'text': "Temperature", 'font': {'size': 24}},
        gauge = {
        'axis': {'range': [31, 33], 'tickwidth': 1, 'tickcolor': "darkblue"},
        'bar': {'color': "black"},
        'bgcolor': "black",
        'borderwidth': 1,
        'bordercolor': "white",
        'steps': [
            {'range': [31, 31.5], 'color': '#cce6ff'},
            {'range': [31.5, 32], 'color': '#80bfff'},
            {'range': [32, 32.5], 'color': '#3399ff'},
            {'range': [32.5, 33], 'color': '#0066cc'},],
        'threshold': {
            'line': {'color': "red", 'width': 4},
            'thickness': 0.75,
            'value': 32.5}}))

    fig.update_layout(paper_bgcolor = "black", font = {'color': "white", 'family': "Arial"},
                     
                     margin=dict(l=20, r=20, t=20, b=20,pad=10),
                     height=110)
    
  
    
    return fig    

@app.callback(
    Output('ghum-graph', 'figure'),
    [ Input('graph-update', 'n_intervals') ]
)
  
def update_graph_scatter(n):

    Y = df["Humidity"][n]
    
    fig = go.Figure(go.Indicator(
    mode = "gauge+number",
    value = Y,
    domain = {'x': [0, 1], 'y': [0, 1]},
    title = {'text': "Humidity", 'font': {'size': 24}},
        gauge = {
        'axis': {'range': [68, 72], 'tickwidth': 1, 'tickcolor': "darkblue"},
        'bar': {'color': "black"},
        'bgcolor': "black",
        'borderwidth': 1,
        'bordercolor': "white",
        'steps': [
            {'range': [68,69], 'color': '#ffcccc'},
            {'range': [69,70], 'color': '#ff8080'},
            {'range': [70,71], 'color': '#ff3333'},
            {'range': [71,72], 'color': '#ff0000'},],
        'threshold': {
            'line': {'color': "blue", 'width': 4},
            'thickness': 0.75,
            'value': 71}}))

    fig.update_layout(paper_bgcolor = "black", font = {'color': "white", 'family': "Arial"},
                     
                     margin=dict(l=20, r=20, t=20, b=20,pad=10),
                     height=110)
    
  
    
    return fig    

@app.callback(
    Output('gco2-graph', 'figure'),
    [ Input('graph-update', 'n_intervals') ]
)
  
def update_graph_scatter(n):

    Y = df["CO2"][n]
    
    fig = go.Figure(go.Indicator(
    mode = "gauge+number",
    value = Y,
    domain = {'x': [0, 1], 'y': [0, 1]},
    title = {'text': "CO2", 'font': {'size': 24}},
        gauge = {
        'axis': {'range': [400, 900], 'tickwidth': 1, 'tickcolor': "darkblue"},
        'bar': {'color': "black"},
        'bgcolor': "black",
        'borderwidth': 1,
        'bordercolor': "white",
        'steps': [
            {'range': [400, 525], 'color': '#ffff99'},
            {'range': [525, 650], 'color': '#ffff4d'},
            {'range': [650, 800], 'color': '#ffff33'},
            {'range': [800, 900], 'color': '#ffff00'},],
        'threshold': {
            'line': {'color': "red", 'width': 4},
            'thickness': 0.75,
            'value': 800}}))

    fig.update_layout(paper_bgcolor = "black", font = {'color': "white", 'family': "Arial"},
                     
                     margin=dict(l=20, r=20, t=20, b=20,pad=10),
                     height=110)
    
  
    
    return fig    

@app.callback(
    Output('gmisc-graph', 'figure'),
    [ Input('graph-update', 'n_intervals') ]
)
  
def update_graph_scatter(n):

    Y = df["Misc"][n]
    
    fig = go.Figure(go.Indicator(
    mode = "gauge+number",
    value = Y,
    domain = {'x': [0, 1], 'y': [0, 1]},
    title = {'text': "CO2", 'font': {'size': 24}},
        gauge = {
        'axis': {'range': [400, 650], 'tickwidth': 1, 'tickcolor': "darkblue"},
        'bar': {'color': "black"},
        'bgcolor': "black",
        'borderwidth': 1,
        'bordercolor': "white",
        'steps': [
            {'range': [400, 462.5], 'color': '#ccffcc'},
            {'range': [462.5, 525], 'color': '#66ff66'},
            {'range': [525, 587.5], 'color': '#1aff1a'},
            {'range': [587.5, 650], 'color': '#009900'},],
        'threshold': {
            'line': {'color': "red", 'width': 4},
            'thickness': 0.75,
            'value': 600}}))

    fig.update_layout(paper_bgcolor = "black", font = {'color': "white", 'family': "Arial"},
                     
                     margin=dict(l=20, r=20, t=20, b=20,pad=10),
                     height=110)
    
  
    
    return fig    



if __name__ == '__main__':
    app.run_server()
```
